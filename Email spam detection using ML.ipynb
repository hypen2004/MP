```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #split the data in training and testing model
from sklearn.feature_extraction.text import TfidfVectorizer #feature of extraction
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mail_data.csv')
```

```python
print(df)
```

```python
#convert mail data into notnull data
data = df.where((pd.notnull(df)),'')
```

```python
data.head(10) #top 10 dataset
```

```python
data.info() #for information
```

```python
#how many rows and columns are there in table
data.shape
```

```python
data.loc[data['Category'] == 'spam', 'Category',] = 0
data.loc[data['Category'] == 'ham', 'Category',] = 1
```

```python
X = data['Message']

Y = data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
X_train
```

```python
Y_train
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python

# transform the text data to feature vectors that can be used as input to the Logistic regression
# Changed 'True' to True
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features) # Assign the predictions to prediction_on_test_data
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data) # Now prediction_on_test_data is defined
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
input_your_mail = ['I ve been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times']

input_data_features = feature_extraction.transform(input_your_mail)

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
